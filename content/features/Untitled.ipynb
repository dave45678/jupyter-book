```python
for ii, ifile in enumerate(html):
    with open(ifile, 'r') as ff:
        lines = ff.readlines()
    for jj, iline in enumerate(lines):
        replaced = False
        if '<link rel="canonical"' in iline:
            link = iline.split('href="')[-1].split('"')[0]
        if 'http-equiv="refresh"' in iline:
            lines[jj] = iline.replace("https://predictablynoisy.com/jupyter-book/", "https://jupyter.org/jupyter-book/")
            replaced = True
    redirect = '<meta http-equiv="refresh" content="0; url={}">'.format(
        link.replace("https://predictablynoisy.com/jupyter-book/", "https://jupyter.org/jupyter-book/"))
    lines.insert(6, redirect)
```
